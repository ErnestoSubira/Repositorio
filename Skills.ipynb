### The following code generates Fibonaccci sequence for the first 20 numbers

```python
# In Fibonacci sequence each number is the result of the sum of two previous numbers. First number is always 0 and Second number is always 1
n = 20
first = 0
second = 1
sum = 0
count = 1
print("Fibonacci(n) sequence for n=20")
while(count<=n):
    print(sum)
    count = count + 1
    first = second
    second = sum
    sum = first + second
```
